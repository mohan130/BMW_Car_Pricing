```python
def func(x):
    if(g.loc[x][0]<=8800):
        return 0
    elif((g.loc[x][0]>8800) and (g.loc[x][0]<=12700)):
        return 1
    elif(g.loc[x][0]>12700 and g.loc[x][0]<=16673):
        return 2
    elif(g.loc[x][0]>16673 and g.loc[x][0]<=24834):
        return 3
    elif(g.loc[x][0]>24834 and g.loc[x][0]<=39064):
        return 4
    elif(g.loc[x][0]>39064 and g.loc[x][0]<=41600):
        return 5
    elif(g.loc[x][0]>41600 and g.loc[x][0]<=49534):
        return 6
    elif(g.loc[x][0]>49534 and g.loc[x][0]<=51050):
        return 6
    elif(g.loc[x][0]>51050 and g.loc[x][0]<=61650):
        return 7
    else:
        return 8
df['model_key_e'] = df.model_key.apply(lambda x: func(x))
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression, Ridge
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_df.drop('price', axis=1),final_df['price'],test_size=0.2, random_state = 42)
```

```python
RFModel = RandomForestRegressor().fit(X_train,y_train)
```

```python
df_feature_imp = pd.DataFrame({
    'Features_imp': RFModel.feature_importances_,
    'Features': final_df.drop('price', axis=1).columns
}).sort_values(by='Features_imp').reset_index(drop=True)
fig, ax = plt.subplots(1,1, figsize= (15,8))
sns.barplot(df_feature_imp.Features_imp,df_feature_imp.Features, ax =ax);
```

```python
from sklearn.metrics import mean_squared_log_error,r2_score
```

```python
print("Train R^2_score {}".format(r2_score(y_train,RFModel.predict(X_train))))
print("Test R^2 score {}".format(r2_score(y_test,RFModel.predict(X_test))))
```

```python
print("Train mean_squared_log_error_score {}".format(mean_squared_log_error(y_train,RFModel.predict(X_train))))
print("Test mean_squared_log_error score {}".format(mean_squared_log_error(y_test,RFModel.predict(X_test))))
```

#### Performing scaling in the features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train, y_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
for model in [RandomForestRegressor(),SVR(),LinearRegression(),Ridge()]:
    scores = cross_val_score(model, X_train, y_train, cv=5,scoring='r2',n_jobs=-1)
    print('Model ==> {}'.format(str(model)))
    print('Avg R^2 {}'.format(np.mean(scores)))
```
